```python
import pandas as pd
```

```python
train_data = pd.read_csv("customer_churn_dataset-training-master.csv")
test_data = pd.read_csv("customer_churn_dataset-testing-master.csv")
```

```python
print(train_data.head())
print(train_data.isnull().sum())
```

```python
train_data.columns = train_data.columns.str.replace(' ', '_')
print(train_data.head())
```

```python
test_data.columns = test_data.columns.str.replace(' ', '_')
print(test_data.head())
```

```python
# Fill or drop missing values as needed
train_data.fillna(method='ffill', inplace=True)
test_data.fillna(method='ffill', inplace=True)
```

```python
if 'Contract_Length' in train_data.columns:
    print("Unique values in 'Contract_Length':", train_data['Contract_Length'].unique())

    # Convert 'Contract_Length' to string if it's not already
    train_data['Contract_Length'] = train_data['Contract_Length'].astype(str)

    # Now extract numeric values
    train_data['Contract_Length'] = train_data['Contract_Length'].str.extract(r'(\d+)')

    # Convert to numeric and handle NaN values
    train_data['Contract_Length'] = pd.to_numeric(train_data['Contract_Length'], errors='coerce')
    train_data['Contract_Length'].fillna(0, inplace=True)
```

```python
# Example: Encoding 'Gender' and 'Subscription Type'
train_data['Gender'] = train_data['Gender'].map({'Male': 0, 'Female': 1})
train_data = pd.get_dummies(train_data, columns=['Subscription_Type'], drop_first=True)
```

```python
X_train = train_data.drop(['CustomerID', 'Churn'], axis=1)  # Assuming 'Churn' is the target column
y_train = train_data['Churn']
```

```python
from sklearn.model_selection import train_test_split

X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(X_train, y_train, test_size=0.2)
```

```python
from lightgbm import LGBMClassifier

model = LGBMClassifier()
model.fit(X_train_split, y_train_split)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_val_split)
print(f"Accuracy: {accuracy_score(y_val_split, y_pred)}")
```

```python
X_test = test_data.drop(['CustomerID'], axis=1)  # Exclude CustomerID if not needed for prediction
```

```python
if 'Contract_Length' in X_test.columns:
    X_test['Contract_Length'] = X_test['Contract_Length'].astype(str)
    X_test['Contract_Length'] = X_test['Contract_Length'].str.extract(r'(\d+)')
    X_test['Contract_Length'] = pd.to_numeric(X_test['Contract_Length'], errors='coerce')
    X_test['Contract_Length'].fillna(0, inplace=True)  # Fill NaN with 0 or choose another strategy

# Process 'Gender' in test data
if 'Gender' in X_test.columns:
    X_test['Gender'] = X_test['Gender'].map({'Male': 0, 'Female': 1})

# One-hot encode 'Subscription Type' in test data
if 'Subscription Type' in X_test.columns:
    X_test = pd.get_dummies(X_test, columns=['Subscription Type'], drop_first=True)
```

```python
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)

# Make predictions on test data
test_predictions = model.predict(X_test)
```

```python
output_df = test_data[['CustomerID']].copy()
output_df['Predicted Churn'] = test_predictions
output_df.to_csv('predicted_churn.csv', index=False)
```
